```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-5, 5, 400)

plt.plot(x, sigmoid(x), color='blue')
plt.title('Sigmoid Function')
plt.grid()
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

```python
plt.plot(x, relu(x), color='red')
plt.title('ReLU Function')
plt.grid()
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.show()
```

```python
w1, b1 = 1.4, 0.5
m = w1 * x + b1

w2, b2 = 0.5, 1.2
n = w2 * m + b2


plt.plot(x, n, color='blue')
plt.title('Only weights and biases')
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

```python
y = relu(x)

plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
weight = 0.8
bias = -3
y1 = weight * x + bias
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y1')
plt.grid()
plt.show()
```

```python
y2 = relu(y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y2')
plt.grid()
plt.show()
```

```python
z1 = 0.2 * x - 0.1
z2 = relu(z1)
plt.plot(x, z2)
plt.xlabel('x')
plt.ylabel('z2')
plt.grid()
plt.show()
```

```python
out = 1 * y2 + 2 * z2 - 0.4
plt.plot(x, relu(out))
plt.xlabel('x')
plt.ylabel('out')
plt.grid()
plt.show()
```

```python
def complex_relu_network(x):
    """
    Create a highly non-linear function using many shifted and scaled ReLU functions
    """
    # Define multiple ReLU functions with different weights, biases, and shifts
    relu_funcs = [
        # (weight, bias, shift)
        (2.0, 1.0, -2.0),    # First ReLU
        (-1.5, -0.5, 1.0),   # Second ReLU with negative weight
        (3.0, 0.5, 0.5),     # Third ReLU
        (-0.8, -1.0, -1.5),  # Fourth ReLU
        (1.2, 0.3, 2.0),     # Fifth ReLU
        (2.5, -0.7, -0.8),   # Sixth ReLU
        (-1.8, 1.2, 3.0),    # Seventh ReLU
        (0.6, -1.5, -3.2),   # Eighth ReLU
        (4.0, 0.2, 1.5),     # Ninth ReLU
        (-2.2, 0.9, -1.0)    # Tenth ReLU
    ]
    
    # Compute sum of transformed ReLUs
    output = sum(
        weight * max(0, weight * (x - shift) + bias) 
        for weight, bias, shift in relu_funcs
    )
    
    return output

# Generate input data
x = np.linspace(-5, 5, 300)

# Compute network output
y = [complex_relu_network(xi) for xi in x]

# Plotting
plt.plot(x, y)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output (y)', fontsize=12)
plt.grid()
plt.tight_layout()
plt.show()
```
